1- Once boş bir Sözlük oluştur

```python
invertory = {}
```

2- İlk ürün detaylarını bir değişkene kaydedin.

```python
ProductNo1 = "Mobil phone"
ProductNo1_quantity = 5
ProductNo1_price = 20000
ProductNo1_releaseYear = 2020
```

3- Envantere detayları ekleyin.

```python
invertory["ProductNo1"]= ProductNo1
invertory["ProductNo1_quantity"] = ProductNo1_quantity
invertory["ProductNo1_price"] = ProductNo1_price
invertory["ProductNo1_releaseYear"] = ProductNo1_releaseYear
```

4- İkinci ürünün detaylarını bir değişkene kaydedin.

```python
ProductNo2 = "Laptop"
ProductNo2_quantity = 10
ProductNo2_price = 50000
ProductNo2_releaseYear = 2023 
```

5- Ürün detaylarını envantere ekleyin.

```python
invertory["ProductNo2"]= ProductNo2
invertory["ProductNo2_quantity"] = ProductNo2_quantity
invertory["ProductNo2_price"] = ProductNo2_price
invertory["ProductNo2_releaseYear"] = ProductNo2_releaseYear
```

6- ProductNo1_releaseYear ve ProductNo2_releaseYear ürünlerinin envanterde olup olmadığını kontrol edin.

```python
print("ProductNo1_releaseYear" in invertory)
print("ProductNo2_releaseYear" in invertory)
```

7- Envanterden her iki ürünün de çıkış yılını silin.

```python
del(invertory["ProductNo1_releaseYear"])
del(invertory["ProductNo2_releaseYear"])
```
